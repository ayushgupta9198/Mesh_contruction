```python
import cv2
import os
from PIL import Image
import matplotlib.pyplot as plt
import glob
import numpy as np
from stl import mesh
for image in glob.glob("/content/drive/MyDrive/Ayush_Gupta/2D/ibtihaj/Mesh_Created/bg_remover/mask/*.png"):
    img=cv2.imread(image)

    l,m=np.where(np.all(img > [5,5,5], axis=2))
    image=os.path.basename(image)
    image=os.path.splitext(image)[0]
    imag="/content/drive/MyDrive/Ayush_Gupta/2D/ibtihaj/Mesh_Created/bg_remover/input_image/"+image+".png"
    grey_img = Image.open(imag).convert('L')
    (ncols,nrows)=grey_img.size
    max_size=(ncols,nrows)
    max_height=10
    min_height=0
    
    grey_img.thumbnail(max_size)
    imageNp = np.array(grey_img)
    maxPix=imageNp.max()
    minPix=imageNp.min()


    # print(imageNp)
    # print(ncols,nrows)
    (ncols,nrows)=grey_img.size
    vertices=np.zeros((nrows,ncols,3))
    # vertices1=np.zeros((nrows,ncols,3))

    for x in range(min(m), ncols):
        for y in range(min(l), nrows):
            pixelIntensity = imageNp[y][x]  #for texture mesh uncomment line 35,36 and comment line 45 
            z = (pixelIntensity * max_height) / maxPix
        # b=z
        # print("ibtehaj",z)
        # if (z!=0):
        #   b=z
        # else:
        #   z=b
        # print("ali",z)
        #print(imageNp[y][x])
            # z=0
        # b=0
            vertices[y][x]=(x, y, z)
        # vertices1[y][x]=(x, y, -z)


    faces=[]

    for x in range(min(m),max(m)-5):
        s=np.where(m==x)
        s=s[0]
        print(x)
        for h in set(s):
            y=l[h]
        
        # create face 1
            if(y<=max(l)-5):
                vertice1 = vertices[y][x]
              # print(y)
                vertice2 = vertices[y][x+1]
                vertice3 = vertices[y+1][x+1]
              # vertice4 = vertices[y][x+1]
                face1 = np.array([vertice1,vertice2,vertice3])

            # create face 2 
                vertice1 = vertices[y][x]
                vertice2 = vertices[y+1][x]
                vertice3 = vertices[y+1][x+1]
                  # vertice4 = vertices[y+1][x]
                face2 = np.array([vertice1,vertice2,vertice3])
                  #create  face 3

                faces.append(face1)
                faces.append(face2)


    print(f"number of faces: {len(faces)}")
    facesNp = np.array(faces)
    # Create the mesh
    # print(facesNp.shape[0])
    surface = mesh.Mesh(np.zeros(facesNp.shape[0] , dtype=mesh.Mesh.dtype))
    for i, f in enumerate(faces):
        for j in range(3):
              surface.vectors[i][j] = facesNp[i][j]
    name=os.path.basename(image)
    name=os.path.splitext(name)[0]
    # Write the mesh to file "cube.stl"
    surface.save("/content/drive/MyDrive/Ayush_Gupta/2D/ibtihaj/Mesh_Created/bg_remover/output/"+ name+".stl")
    # print(surface)
```
